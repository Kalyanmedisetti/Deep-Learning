```python
from ultralytics import YOLO

# Load the trained model
model = YOLO("runs/detect/train/weights/last.pt")

# Evaluate the model on validation data
metrics = model.val()

# Print the results
print(metrics)
```

```python
''''Box(P) – Precision: Measures how many of the predicted tumor detections were actually correct. A precision of 0.461 (46.1%) means that about 46.1% of the detected tumors were true positives.

R – Recall: Measures how many actual tumors the model successfully detected. A recall of 0.793 (79.3%) means that the model identified 79.3% of actual tumors present in the dataset.

mAP50 – Mean Average Precision at IoU=0.50: Measures the model’s overall detection accuracy. Here, mAP50 = 0.479 (47.9%), which means that the model performs moderately well in localizing tumors but has room for improvement.

mAP50-95 – Mean Average Precision across multiple IoU thresholds: Provides a stricter measure of accuracy. Here, mAP50-95 = 0.358 (35.8%), indicating that the model struggles with precise localization.
```
